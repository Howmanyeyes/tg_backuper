## Списки

Вам дан список. Создайте новый список, в котором порядок следования элементов будет обратным относительно исходного списка.

```python
# ваше решение
a = [1,2,3,4]
b = a[::-1]
print(b)
```

Вам дан список. Создайте новый список, в котором будут присутствовать только чётные элементы исходного списка.

```python
# ваше решение
a = [1, 2, 3, 4]
a = [x for x in a if x%2 == 0]
a
```

```python
# 4.
import seaborn as sns
# df = df.drop(columns='SpType')
df = pd.read_csv("Stars_balanced.csv")
fd = df.drop(columns='SpType') # Дроп т.к. данные в этом стб не числовые
sns.heatmap(fd.corr(), annot=True)
plt.show()
```

```python
# 5.
sns.pairplot(df, hue='TargetClass')
plt.show()
```

```python
# 6.
for column in fd.columns:
      plt.figure(figsize=(10, 5))
      sns.boxplot(x=df[column])
      plt.title(f'Boxplot для переменной {column}')
      plt.show()
      Q1 = fd[column].quantile(0.25)
      Q3 = fd[column].quantile(0.75)
      IQR = Q3 - Q1
      outliers = fd[(fd[column] < (Q1 - 1.5 * IQR)) | (fd[column] > (Q3 + 1.5 * IQR))]
      if not outliers.empty:
          print(f"\nВыбросы в колонке {column}:")
          print(outliers)

```

```python
# 7.
for column in fd.columns:
      print(f"\nСтатистические показатели для класса {column}:\n")
      print("Среднее значение:", fd[column].mean())
      print("Медиана:", fd[column].median())
      print("Максимум:", fd[column].max())
      print("Минимум:", fd[column].min())
```
